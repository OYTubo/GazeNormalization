```python
import h5py
import warp_norm
import data_loader
from config import get_config
```

```python
config, unparsed = get_config()
config.is_train = False
config.batch_size = 100
config.use_gpu = True
config.ckpt_dir = './ckpt'
config.pre_trained_model_path = './ckpt/epoch_24_ckpt.pth.tar'
test_data = data_loader.get_pre_loader('/home/hgh/hghData/Datasets',batch_size=config.batch_size,num_workers=0)
```

```python
# 创建 HDF5 文件
hdf5_path = '/home/hgh/hghData/Datasets/dataset.h5'
hdf5_file = h5py.File(hdf5_path, 'w')


# 遍历 DataLoader，每100张图像为一个 group 存储
for batch_idx, (preprocessed_images, labels, image_paths) in enumerate(test_data):
    start_idx = batch_idx * 100
    end_idx = start_idx + len(image_paths)

    # 在 HDF5 文件中创建一个 group
    group_name = f'group_{batch_idx}'
    group = hdf5_file.create_group(group_name)

    # 将原始图像路径、预处理图像和标签写入 group
    group.create_dataset('image_paths', data=image_paths)
    print(image_paths)
    group.create_dataset('preprocessed_images', data=preprocessed_images)
    group.create_dataset('labels', data=labels)

# 关闭 HDF5 文件
hdf5_file.close()
        
```
